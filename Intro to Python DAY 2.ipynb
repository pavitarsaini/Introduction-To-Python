## Variables

Variables are used to attach a name to a value.
In this case  we are attaching the string `'Hello, World'` to `message`

Now the name has no quotes around it, and as I said earlier this means that Python tries to interpret it as something it should do something with. What it does is to look up the name and substitute in the attached value.
Whenever the name is used, Python will look it up and substitute in the attached value. 


```python
message = 'Hello, world!'
print(message)
```

## Adding variables together

Just like strings, we can add variables together when printing.

```python
message = 'Hello, '
message2 = 'world!'
print(message + message2)
```
