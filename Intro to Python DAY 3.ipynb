We need to be able to compare numbers.

## Boolean
True and False are (the only) two values of a special Python type called a “Boolean” used for recording whether something is true or not.


### Equality Comparison

Used to compare two values
`value1 == value2`

```python
1 == 1
```

### Textual Comparisons

Alphabetic ordering

```python
'cat' < 'dog'
```

Uppercase before lowercase

```python
'Cat' < 'cat'
```

All uppercase before lowercase

```python
'DOG' < 'cat'
```

### Inequalities 

A common requirement is to determine if a number lies in a particular range. For this purpose, Python supports the mathematical notation a < b < c. The inequalities can be any combination that make sense.


```python
0 < 5 < 10
```

### Converting To Booleans


As with all Python types there is a function named after the type that tries to convert arbitrary inputs into Booleans. Given that there are only two Boolean values this tends to be a very simple function.

Empty string is `False`


```python
bool('')
```

Non-empty string is `True`

```python
bool('Fred')
```

Zero is `False`

```python
bool(0)
```

Non-zero is `True`


```python
bool(12)
```

# Boolean Operations


### `and`
 The and of two booleans values is True if (and only if) both its inputs are True. If either is False then its output is Fal

```python
4 < 5 and 6 < 7
```

```python
4 < 5 and 6 > 7
```

### `or`
The results of this operation is True if either of its inputs are True and False only if both its inputs are False.


```python
4 < 5 or 6 < 7
```

```python
4 < 5 or 6 > 7
```

### `not`
This takes only one input and “flips” it. True becomes False and vice versa.



```python
not 6 < 7
```

```python
not 6 > 7
```
