## Boolean Operations


### `and`
 The and of two booleans values is True if (and only if) both its inputs are True. If either is False then its output is Fal

```python
4 < 5 and 6 < 7
```

### `or`
The results of this operation is True if either of its inputs are True and False only if both its inputs are False.


```python
4 < 5 or 6 < 7
```

### `not`
This takes only one input and “flips” it. True becomes False and vice versa.



```python
not 6 < 7
```

## Assignment


### Simple Evaluations

RHS is evaluated first. Then Python attachs the LHS to it.

```python
beta = 100 + 20
print(beta)
```

### Changing Values
Python lets you set a name for a value and then change it.


```python
gamma = 1
gamma = 2
print(gamma)
```

### Name on the RHS


1.   Python recognizes the three tokens on the RHS. These are the name “alpha” the “+” and the literal integer 40.
2.   Looks up the names. 
3.   Runs the function to give an integer 140 object in memory.


```python
alpha = 100
delta = alpha + 40
print(delta)
```

### Same name on both sides


```python
alpha = 100
delta = alpha + 40
delta = delta + 10
print(delta)
```

**Short-cut**
The operation of modifying the value attached to a name is so common that Python, and some other languages, have short-cuts in their syntax to make the operations shorter to write. These operations are called “augmented assignments”.


```python
alpha = 100
delta = alpha + 40
delta += 10
print(delta)
```

# **Looping**





## While
The first type of loop is the `while` loop.
With the `while` loop we can execute a set of statements as long as a condition is true.

```python
number = 1

while number <= 10 :
  print(number)
  number += 1

print('Done!')
```

# **If**
Decision making is required when we want to execute a code only if a certain condition is satisfied.

The `if` statement is used in Python for decision making.

```python
text = input('Number? ')
number = int(text)

if number % 2 == 0 :
    print('Even number')
else :
    print('Odd number')

print('That was fun!')
```

# Input validation
An important step is “input validation”, where we check that the input makes sense. 

```python
text = input('Number? ')
number = float(text)

if number < 0.0:
  print('Number must be positive!')
exit()
```

# “Chained” tests
The input validation phase comes straight after the input itself.

```python
text = input('Number?␣')
number = int(text)

if number == 0:
  print('Number cannot be zero!')
  exit()

elif number < 0:
  print('Number must be positive!')
elif number >= 1:
  print('Got it!')
else:
  print('Number not valid!')
```
