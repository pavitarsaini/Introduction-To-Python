## Adding strings together.

This is called **“Concatenation”**.

```python
print('Hello, ' + 'world!')
```

Why we need different quotes

```python
#This does not work
print('He said 'hello' to her.')
```

```python
#This does work
print('He said "hello" to her.')
```

Adding arbitrary quotes

```python
#This also works
print('He said \'hello\' to her.')
```

## Line breaks in text

You cannot hit enter and get a line break. Try bellow. 

```python
#This does not work
print('Hello 
world!')
```

You have to use an **Escape Character** "/n" is for a line break

```python
#This does work
print('Hello \nWorld!')
```
